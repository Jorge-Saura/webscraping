```python
soup = BeautifulSoup('<b id="boldest" class="boldest">Extremely bold</b>')

tagb=soup.b
print(type(tagb))

print(tagb.name)

print(tagb.attrs)

print(tagb['id'])

#modificar
tagb['id']='verybold'
tagb['new-attribute']='new'

print(tagb)

del tagb['new-attribute']
print(tagb)

#consultar contenido
print(tagb.string)
print(type(tagb.string))
tagb.string.replace_with("not longer bold")
print(tagb)
```
